## 1 - Criando uma lista com os nomes de todos os municípios do Ceará

````python
# Site com todos os municípios do Ceará
lista_munic = "https://www.ipece.ce.gov.br/consulta-aos-mapas-e-legislacao/"
````

````python
# Fazendo a requisição do site por meio do Requests
resposta = requests.get(lista_munic)
````

````python
print(resposta)
````

````python
# Parsear o HTML da página com a biblioteca BeautifulSoup
municipios = BeautifulSoup(resposta.text, 'html.parser') 
````

````python
# Mostrando o HTML organizado
print(municipios.prettify())
````

````python
# Selecionando a tag da página HTML que contém o conteúdo que queremos
municipios = municipios.find_all("strong")
municipios
````

````python
municipios[0].contents
````

````python
len(municipios)
````

````python
# Fazendo um for para pegarmos todos os nomes dos municípios e armazená-los na variável munic_ceara
munic_ceara = []
for i in range(len(municipios)):
    munic_ceara.append(municipios[i].text)
````

````python
print(munic_ceara)
len(munic_ceara)
````

**Observação:** para a criação da lista com a URL de cada município será necessário removermos os acentos e substituir os espaços em branco por hífens. 

````python
# remover acentos de cada município na lista
municipios_sem_acentos = [unidecode(municipio) for municipio in munic_ceara]

print(municipios_sem_acentos)
len(municipios_sem_acentos)
````

````python
municipios_sem_espacos = [municipio.replace(' ', '') if ' ' in municipio else municipio for municipio in municipios_sem_acentos]

print(municipios_sem_espacos)
````

````python
municipios_com_hifen = [municipio.replace(' ', '-') if ' ' in municipio else municipio for municipio in municipios_sem_acentos]
print(municipios_com_hifen)
````

## 3 - Fazendo o scraping

------------
A estrutura do html onde esta os valores que eu quero fazer o scrapy, é a seguinte:

```
<div>
    <div class="ind-label">
      <img alt="Área Territorial" >
      <p>Área Territorial</p>
    </div>
    <p class="ind-value">
      text
    </p>
</div>
```
Para eu poder conseguir pegar o texto que esta na segunda tag 'p', eu vou procurar por uma tag 'img', e pegar a tag antecessora a ela usando o metodo '.find_previous()', mais ainda não é a tag que eu quero, então vou usar novamento o metodo 'find_previous()', agora a tag que vai ser retornada vai ser a primeira div (a div pai), pronto agora é só pegar o texto da tag 'p' 



````python
headers = { 
    'User-Agent'      : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36', 
    'Accept'          : 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8', 
    'Accept-Language' : 'en-US,en;q=0.5',
    'DNT'             : '1', # Do Not Track Request Header 
    'Connection'      : 'close'
}

results_list = []
counter = 1
````

````python
for url in urls:
  try:
    data = requests.get(url, headers=headers, timeout=5).text
    print(f'Requisição {counter}: {url}')
    soup = BeautifulSoup(data,"html.parser")
    

    # Vamos filtrar os valores desejados e salvá-los detro de um dicionario que vai ser salvo na lista results_list
    codigo = soup.find('div',class_='quick-facts-titulo').find('p',class_='codigo').text.split(' ')[1]
    cidade = soup.find('div',class_='quick-facts-titulo').find('h1').text
    prefeito = soup.find('p', text="Prefeito").find_previous().find_previous('div').find('p',class_='ind-value').text.split('\xa0\xa0\xa0')[0]
    gentilico = soup.find('p', text="Gentílico").find_previous().find_previous('div').find('p',class_='ind-value').text.split('\xa0\xa0\xa0')[0]
    area_territorial = soup.find('img', alt='Área Territorial').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    
    
    populacao_estimada = soup.find('img', alt='População estimada').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    densidade_demografica = soup.find('img', alt='Densidade demográfica').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    escolarizacao = soup.find('img', alt='Escolarização <span>6 a 14 anos</span>').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    idhm = soup.find('img', alt='IDHM <span>Índice de desenvolvimento humano municipal</span>').find_previous().find_previous('div').find('p',class_='ind-value').text[:5]
    mortalidade_infantil = soup.find('img', alt='Mortalidade infantil').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    receitas_realizadas = soup.find('img', alt='Receitas realizadas').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    despesas_empenhadas = soup.find('img', alt='Despesas empenhadas').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    pib = soup.find('img', alt='PIB per capita').find_previous().find_previous('div').find('p',class_='ind-value').text.split(' ')[0]
    
    cidade_x = {
        'codigo': codigo,
        'cidade': cidade,
        'prefeito': prefeito,
        'gentilico': gentilico,
        'area_territorial': area_territorial,
        'populacao_estimada': populacao_estimada,
        'densidade_demografica': densidade_demografica,
        'escolarizacao': escolarizacao,
        'idhm': idhm,
        'mortalidade_infantil': mortalidade_infantil,
        'receitas_realizadas': receitas_realizadas,
        'despesas_empenhadas': despesas_empenhadas,
        'pib': pib
        }
    results_list.append(cidade_x)

    counter += 1
    #Vai espera um tempo aleatório entre 4 e 5 segundos antes de fazer a próxima requisição
    time.sleep(random.uniform(4, 5))
  except requests.exceptions.RequestException as e:
        print("Erro ao fazer a requisição:", e)
        counter += 1
````

````python
#Trasformado a lista 'results_list' em um dataframe
df_cidade = pd.DataFrame(results_list)
df_cidade.head()
````

````python
#Salvado o dataframe
df_cidade.to_csv('DF_Cidades.csv',index=False)
````
